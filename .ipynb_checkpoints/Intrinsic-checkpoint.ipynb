```python
intrinsic_dir = "C:/Users/Keyan/Desktop/Science/Data/P-Cygni/Data/intrinsic_pols/"
files = glob(intrinsic_dir+"*")
retfiles = [f    for f in files if f[69]=='o']
rfiles = [f    for f in files if f[69]=='r']
bfiles = [f    for f in files if f[69]=='b']
matched = []
for rf in rfiles:
    for bf in bfiles:
        if rf[64:69]==bf[64:69]:
            matched.append([rf,bf])
matched = np.array(matched)
mdfs = [[pd.read_csv(f[0]),pd.read_csv(f[1])]    for f in matched]
mdfs = [pd.concat([p[0],p[1][p[1].Wavelength < 6100]]).sort_values("Wavelength")    for p in mdfs]
mdfs = [df[pc.polarization(df.Q,df.U)[0] < 0.8]   for df in mdfs]
rdfs = [pd.read_csv(f)    for f in files if f[72]=='r']
bdfs = [pd.read_csv(f)    for f in files if f[72]=='b']
retdfs = [pd.read_csv(f)     for f in retfiles]
alldfs = retdfs+mdfs
t = [float(f[64:69])    for f in retfiles] + [float(f[0][64:69])    for f in matched]
mt = np.array([float(f[0][64:69])    for f in matched])
mphaset = (mt/97)%1
```

```python
qs = bt.ivq
us = bt.ivu
es = bt.sve
mq,mu = np.array([sum(qs*(es**-2))/sum(es**-2)]),np.array([sum(us*(es**-2))/sum(es**-2)])

fig,ax = plt.subplots(1,figsize=(4,4.2));
ax.set_aspect("equal")
ax.set_ylabel("$u$",fontsize=20,usetex=True,rotation=0)
ax.set_xlabel("$q$",fontsize=20,usetex=True,labelpad=-3)
ax.axhline(0,linestyle='--',color='black',alpha=0.5)
ax.axvline(0,linestyle='--',color='black',alpha=0.5)
plt.scatter(qs,us,c='black')
# Average Error is smaller than the marker size
ax.errorbar(qs,us,linestyle='none',yerr=es,xerr=es,color='black')
# ax.errorbar(np.mean(qs),np.mean(us),linestyle='none',yerr=4*np.sqrt(np.sum(es**2))/np.sqrt(len(es)),xerr=4*np.sqrt(np.sum(es**2))/np.sqrt(len(es)),color='black')
ax.set_ylim(-0.6)
plt.tight_layout(pad=0)
plt.savefig(paperdir+"/intrinsic_QU.png",overwrite=True,dpi=300)
print(pc.polarization(np.mean(qs),np.mean(us)),np.degrees(np.arctan2(np.mean(qs),np.mean(us))))
```

```python
interp_Q = []
interp_U = []
interp_err = []
for i in range(1, len(mdfs)):
    interpolated_Q = np.interp(mdfs[0].Wavelength, mdfs[i].Wavelength, mdfs[i].Q)
    interp_Q.append(interpolated_Q)
    interpolated_U = np.interp(mdfs[0].Wavelength, mdfs[i].Wavelength, mdfs[i].U)
    interp_U.append(interpolated_U)
    interpolated_err = np.interp(mdfs[0].Wavelength, mdfs[i].Wavelength, mdfs[i].Error)
    interp_err.append(interpolated_err)
mean_Q = np.sum(np.array(interp_Q)*(np.array(interp_err)**-2), axis=0)/np.sum(np.array(interp_err)**-2, axis=0)
mean_U = np.sum(np.array(interp_U)*(np.array(interp_err)**-2), axis=0)/np.sum(np.array(interp_err)**-2, axis=0)
mean_err = np.sqrt(np.sum(np.array(interp_err)**2, axis=0) / len(interp_err))
```

```python
mpol,mpos = pc.polarization(mean_Q,mean_U)
mpaerr = pc.position_angle_error(mpol,mean_err)
```

```python
fig,(ax1,ax2,ax3,ax4) = plt.subplots(4,figsize=[4,6],sharex=True)
fig.subplots_adjust(wspace=0.01,hspace=0)

ax1.errorbar(mdfs[0].Wavelength,mpol,yerr=mean_err,color='black')
ax1.set_xlim(6563-80,6563+80)
ax1.set_ylim(-0.1,0.45)
ax1.set_ylabel("$P_\%$",size=20,usetex=True,rotation=0,labelpad=15);

ax2.errorbar(mdfs[0].Wavelength,mpos,yerr=mpaerr,color='black')
ax2.set_xlim(6563-80,6563+80)
ax2.set_ylim(-150,150)
ax2.set_ylabel("$\Theta [^\circ]$",size=20,usetex=True,rotation=0,labelpad=15);

ax3.errorbar(mdfs[0].Wavelength,mean_Q,yerr=mean_err,color='black')
ax3.set_ylim(-0.2,0.2)
ax3.set_ylabel("$q$",size=20,usetex=True,rotation=0);

ax4.errorbar(mdfs[0].Wavelength,mean_U,yerr=mean_err,color='black')
ax4.set_ylim(-0.3,0.4)
ax4.set_xlabel("$\lambda$ [\AA]",size=20,usetex=True,labelpad=0);
ax4.set_ylabel("$u$",size=20,usetex=True,rotation=0);

ax1.axvspan(6540,6600,color='black',alpha=0.1)
ax1.axvline(6540,linestyle='--',color='black')
ax1.axvline(6600,linestyle='--',color='black')
ax2.axvspan(6540,6600,color='black',alpha=0.1)
ax2.axvline(6540,linestyle='--',color='black')
ax2.axvline(6600,linestyle='--',color='black')
ax3.axvspan(6540,6600,color='black',alpha=0.1)
ax3.axvline(6540,linestyle='--',color='black')
ax3.axvline(6600,linestyle='--',color='black')
ax4.axvspan(6540,6600,color='black',alpha=0.1)
ax4.axvline(6540,linestyle='--',color='black')
ax4.axvline(6600,linestyle='--',color='black')

plt.tight_layout(pad=0)
plt.savefig(paperdir+"/halpha.png",overwrite=True,dpi=300)
```

```python
cutdfs = [df[(df.Wavelength > 4000) & (df.Wavelength < 7500)]    for df in alldfs]
interp_Q = []
interp_U = []
interp_err = []
for i in range(1, len(cutdfs)):
    interpolated_Q = np.interp(cutdfs[0].Wavelength, cutdfs[i].Wavelength, cutdfs[i].Q)
    interp_Q.append(interpolated_Q)
    
    interpolated_U = np.interp(cutdfs[0].Wavelength, cutdfs[i].Wavelength, cutdfs[i].U)
    interp_U.append(interpolated_U)
    
    interpolated_err = np.interp(cutdfs[0].Wavelength, cutdfs[i].Wavelength, cutdfs[i].Error)
    interp_err.append(interpolated_err)
mean_Q = np.mean(np.array(interp_Q), axis=0)
mean_U = np.mean(np.array(interp_U), axis=0)
mean_err = np.sqrt(np.sum(np.array(interp_err)**2, axis=0) / len(interp_err))
cutpol,cutpos = pc.polarization(mean_Q,mean_U)
cutpaerr = pc.position_angle_error(cutpol,mean_err)
```

```python
fig,(ax1,ax2,ax3,ax4) = plt.subplots(4,figsize=[4,4],sharex=True)
plt.tight_layout(rect=(0.01,0.02,1,1))
fig.subplots_adjust(wspace=0.01,hspace=0)

ax1.errorbar(cutdfs[0].Wavelength,cutpol,color='black')
ax1.set_ylim(0,0.3)
ax1.errorbar(cutdfs[0].Wavelength,cutpol,yerr=mean_err,alpha=0.075,color='black',linestyle='None')

ax2.errorbar(cutdfs[0].Wavelength,mean_Q,color='black')
# ax2.set_ylim(-0.2,0.25)
ax2.errorbar(cutdfs[0].Wavelength,mean_Q,yerr=mean_err,alpha=0.075,color='black',linestyle='None')

ax3.errorbar(cutdfs[0].Wavelength,mean_U,color='black')
# ax3.set_ylim(-0.2,0.35)
ax3.errorbar(cutdfs[0].Wavelength,mean_U,yerr=mean_err,alpha=0.075,color='black',linestyle='None')

ax4.errorbar(cutdfs[0].Wavelength,cutpos,color='black')
ax4.set_ylim(-100,100)
ax4.set_xlim(4000,7500)

ax1.set_ylabel("$P_\%$",size=20,usetex=True,rotation=0);
ax2.set_ylabel("$q$",size=20,usetex=True,rotation=0);
ax3.set_ylabel("$u$",size=20,usetex=True,rotation=0);
ax4.set_xlabel("$\lambda$ [\AA]",size=20,usetex=True);
ax4.set_ylabel("$\Theta$ [$^\circ$]",size=20,usetex=True,rotation=0,verticalalignment='center',labelpad=5)

plt.tight_layout(pad=0)
plt.savefig(paperdir+"/intmean.png",dpi=300,overwrite=True)
```

```python
line = lambda x,m,b: m*x+b
ms = []
mdevs = []
for df in cutdfs:
    pol,pos = pc.polarization(df.Q,df.U)
    (m,b),cov = curve_fit(line,df.Wavelength,pol,sigma=df.Error,p0=(0,np.mean(pol)))
    mdevs.append(np.sqrt(np.diag(cov)[0]))
    ms.append(m)
fig,ax = plt.subplots(1,figsize=[8,4])

ax.axhline(0,linestyle='--',alpha=0.3)
ax.axvline(49750,linestyle='--',alpha=0.6)
ax.errorbar(bt.MJD,np.array(ms),yerr=np.array(mdevs),linestyle='none',elinewidth=2,capsize=4,color='black')

ax.set_xlabel("MJD",size=20,usetex=True);
ax.set_ylabel(r"\%\AA$^{-1}$",size=20,usetex=True,rotation=0,labelpad=20);
plt.tight_layout(pad=0)
plt.savefig(paperdir+"/slope.png",dpi = 300, overwrite=True)
```

```python
ms = []
mdevs = []
for df in cutdfs:
    pol,pos = pc.polarization(df.Q,df.U)
    (m,b),cov = curve_fit(line,df.Wavelength,pos,sigma=df.Error,p0=(0,np.mean(pos)))
    mdevs.append(np.sqrt(np.diag(cov)[0]))
    ms.append(m)
fig,ax = plt.subplots(1,figsize=[8,4])

ax.axhline(0,linestyle='--',alpha=0.3)
ax.axvline(49750,linestyle='--',alpha=0.6)
print(np.mean(ms))
ax.errorbar(bt.MJD,np.array(ms),yerr=np.array(mdevs),linestyle='none',elinewidth=2,capsize=4,color='black')

ax.set_xlabel("MJD",size=20,usetex=True);
ax.set_ylabel(r"degree\AA$^{-1}$",size=20,usetex=True,rotation=0,labelpad=20);
plt.tight_layout(pad=0)
plt.savefig(paperdir+"/slopepa.png",dpi = 300, overwrite=True)
```

```python
ls = LombScargle(bt.MJD,ms,dy=mdevs)
print(ls.false_alarm_level(0.01))
freq,power = ls.autopower()
plt.axvline(x=1/64.767)
plt.plot(freq,power)
```
