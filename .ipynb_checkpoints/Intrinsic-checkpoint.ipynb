```python
import P_Cyg as pc
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from glob import glob
from matplotlib import rcParams
from astropy.stats import LombScargle
paperdir = "C:/Users/Keyan/Desktop/Science/Data/P-Cygni/Figures/Paper_Figures"
rcParams['text.usetex'] = True
rcParams['figure.figsize'] = [10,10]
rcParams['lines.color'] = 'black'
rcParams['font.size'] = 20
rcParams['font.family'] = 'serif'
rcParams['legend.fontsize'] = 14
rcParams['xtick.labelsize']= 14
rcParams['ytick.labelsize']= 14
```

```python
intrinsic_dir = "C:/Users/Keyan/Desktop/Science/Data/P-Cygni/Data/intrinsic_pols/"
files = glob(intrinsic_dir+"*")
retfiles = [f    for f in files if f[72]=='o']
rfiles = [f    for f in files if f[72]=='r']
bfiles = [f    for f in files if f[72]=='b']
matched = []
for rf in rfiles:
    for bf in bfiles:
        if rf[64:69]==bf[64:69]:
            matched.append([rf,bf])
matched = np.array(matched)
mdfs = [[pd.read_csv(f[0]),pd.read_csv(f[1])]    for f in matched]
mdfs = [pd.concat([p[0],p[1][p[1].Wavelength < 6100]]).sort_values("Wavelength")    for p in mdfs]
rdfs = [pd.read_csv(f)    for f in files if f[72]=='r']
bdfs = [pd.read_csv(f)    for f in files if f[72]=='b']
retdfs = [pd.read_csv(f)     for f in retfiles]
alldfs = retdfs+mdfs
t = [float(f[64:69])    for f in retfiles] + [float(f[0][64:69])    for f in matched]
mt = np.array([float(f[0][64:69])    for f in matched])
mphaset = (mt/97)%1
```

```python
qs = np.array([np.mean(df.Q[(df.Wavelength > 6700) & (df.Wavelength < 7500)])   for df in alldfs])
us = np.array([np.mean(df.U[(df.Wavelength > 6700) & (df.Wavelength < 7500)])   for df in alldfs])
es = np.array([np.sqrt(np.sum(df.Error[(df.Wavelength > 6700) & (df.Wavelength < 7500)]**2))/
               df.Error[(df.Wavelength > 6700) & (df.Wavelength < 7500)].size    
               for df in alldfs])
fig,ax = plt.subplots(1,figsize=(9.2,10));
plt.tight_layout(rect=(0,0.01,1,1))
ax.set_aspect("equal")
ax.set_ylabel("Stokes U",fontsize=25)
ax.set_xlabel("Stokes Q",fontsize=25)
ax.axhline(0,linestyle='--',color='black',alpha=0.5)
ax.axvline(0,linestyle='--',color='black',alpha=0.5)
plt.scatter(qs,us,c='black')
# Average Error is smaller than the marker size
ax.errorbar(qs,us,linestyle='none',yerr=es,xerr=es,color='black')
# ax.errorbar(np.mean(qs),np.mean(us),linestyle='none',yerr=4*np.sqrt(np.sum(es**2))/np.sqrt(len(es)),xerr=4*np.sqrt(np.sum(es**2))/np.sqrt(len(es)),color='black')
ax.set_ylim(-0.6)
plt.savefig(paperdir+"/intrinsic_QU.png",transparent=True,overwrite=True,dpi=300)
```

```python
phaset = (np.array(t)/97)%1
(pt,pbinq),(pt,pbinu) = pc.easy_bin_mean(phaset,qs,10),pc.easy_bin_mean(phaset,us,10)
fig,ax = plt.subplots(1,figsize=(9.2,10));
plt.tight_layout(rect=(0,0.01,1,1))
ax.set_aspect("equal")
ax.axhline(0)
ax.axvline(0)
ax.set_ylabel("Stokes U",fontsize=25)
ax.set_xlabel("Stokes Q",fontsize=25)
ax.scatter(pbinq,pbinu,color='black',s=150)
ax.plot(pbinq,pbinu,c="black",linestyle='--',alpha=0.6)
plt.savefig(paperdir+"/qutracks.png",overwrite=True,dpi=300)
```

```python
interp_Q = []
interp_U = []
interp_err = []
for i in range(1, len(mdfs)):
    interpolated_Q = np.interp(mdfs[0].Wavelength, mdfs[i].Wavelength, mdfs[i].Q)
    interp_Q.append(interpolated_Q)
    
    interpolated_U = np.interp(mdfs[0].Wavelength, mdfs[i].Wavelength, mdfs[i].U)
    interp_U.append(interpolated_U)
    
    interpolated_err = np.interp(mdfs[0].Wavelength, mdfs[i].Wavelength, mdfs[i].Error)
    interp_err.append(interpolated_err)
mean_Q = np.mean(np.array(interp_Q), axis=0)
mean_U = np.mean(np.array(interp_U), axis=0)
mean_err = np.sqrt(np.sum(np.array(interp_err)**2, axis=0) / len(interp_err))
```

```python
mpol,mpos = pc.polarization(mean_Q,mean_U)
mpaerr = pc.position_angle_error(mpol,mean_err)
```

```python
fig,(ax1,ax2,ax3) = plt.subplots(3,figsize=[10,10],sharex=True)
plt.tight_layout(rect=(0.01,0.02,1,1))
fig.subplots_adjust(wspace=0.01,hspace=0)
ax1.errorbar(mdfs[0].Wavelength[150:-100],mpol[150:-100],yerr=mean_err[150:-100],color='black')
ax1.set_xlim(6563-80,6563+80)
ax1.set_ylim(-0.1,0.45)
ax1.set_yticklabels([None,0,0.1,0.2,0.3,0.4])
ax1.set_ylabel("\% Polarization",size=25,usetex=True);

ax2.errorbar(mdfs[0].Wavelength,mean_Q,yerr=mean_err,color='black')
ax2.set_ylim(-0.25,0.15)
ax2.set_ylabel("$q$",size=25,usetex=True,rotation=0);


ax3.errorbar(mdfs[0].Wavelength,mean_U,yerr=mean_err,color='black')
ax3.set_ylim(-0.3,0.4)
ax3.set_xlabel("Wavelength (\AA)",size=25,usetex=True);
ax3.set_ylabel("$u$",size=25,usetex=True,rotation=0);

ax1.axvspan(6540,6600,color='black',alpha=0.1)
ax1.axvline(6540,linestyle='--',color='black')
ax1.axvline(6600,linestyle='--',color='black')
ax2.axvspan(6540,6600,color='black',alpha=0.1)
ax2.axvline(6540,linestyle='--',color='black')
ax2.axvline(6600,linestyle='--',color='black')
ax3.axvspan(6540,6600,color='black',alpha=0.1)
ax3.axvline(6540,linestyle='--',color='black')
ax3.axvline(6600,linestyle='--',color='black')
plt.savefig(paperdir+"/halpha.png",transparent=True,overwrite=True,dpi=300)
```

```python
cutdfs = [df[(df.Wavelength > 4000) & (df.Wavelength < 7500)]    for df in alldfs]
interp_Q = []
interp_U = []
interp_err = []
for i in range(1, len(cutdfs)):
    interpolated_Q = np.interp(cutdfs[0].Wavelength, cutdfs[i].Wavelength, cutdfs[i].Q)
    interp_Q.append(interpolated_Q)
    
    interpolated_U = np.interp(cutdfs[0].Wavelength, cutdfs[i].Wavelength, cutdfs[i].U)
    interp_U.append(interpolated_U)
    
    interpolated_err = np.interp(cutdfs[0].Wavelength, cutdfs[i].Wavelength, cutdfs[i].Error)
    interp_err.append(interpolated_err)
mean_Q = np.mean(np.array(interp_Q), axis=0)
mean_U = np.mean(np.array(interp_U), axis=0)
mean_err = np.sqrt(np.sum(np.array(interp_err)**2, axis=0) / len(interp_err))
cutpol,cutpos = pc.polarization(mean_Q,mean_U)
cutpaerr = pc.position_angle_error(cutpol,mean_err)
```

```python
fig,(ax1,ax2,ax3,ax4) = plt.subplots(4,figsize=[10,10],sharex=True)
plt.tight_layout(rect=(0.01,0.02,1,1))
fig.subplots_adjust(wspace=0.01,hspace=0)
ax1.errorbar(cutdfs[0].Wavelength,cutpol,color='black')
ax1.set_ylabel("\% Polarization",size=25,usetex=True);
# ax1.set_ylim(0,0.4)
ax1.errorbar(5000,np.mean(cutpol),yerr=np.mean(mean_err),alpha=0.4,color='black',elinewidth=4)

ax2.errorbar(cutdfs[0].Wavelength,mean_Q,color='black')
ax2.set_ylabel("$q$",size=25,usetex=True,rotation=0);
# ax2.set_ylim(-0.2,0.25)
ax2.errorbar(5000,np.mean(mean_Q),yerr=np.mean(mean_err),alpha=0.4,color='black',elinewidth=4)

ax3.errorbar(cutdfs[0].Wavelength,mean_U,color='black')
ax3.set_ylabel("$u$",size=25,usetex=True,rotation=0);
# ax3.set_ylim(-0.2,0.35)
ax3.errorbar(5000,np.mean(mean_U),yerr=np.mean(mean_err),alpha=0.4,color='black',elinewidth=4)

ax4.errorbar(cutdfs[0].Wavelength,cutpos,color='black')
# ax4.set_ylim(-100,100)
ax4.set_xlim(4000,7500)
ax4.set_xlabel("Wavelength [\AA]",size=25,usetex=True);
ax4.set_ylabel("$\Theta$ [$^\circ$]",size=25,usetex=True,rotation=0)
ax4.errorbar(5000,np.mean(cutpos),yerr=np.mean(cutpaerr),alpha=0.4,color='black',elinewidth=4)

plt.savefig(paperdir+"/intmean.png",dpi=300,overwrite=True,transparent=True)
```
