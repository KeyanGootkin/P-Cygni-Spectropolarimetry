```python
% matplotlib inline

import P_Cyg as pc
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from glob import glob
from astropy.io import fits
```

```python
datadir = 'C:/Users/Keyan/Desktop/Science/Data/P-Cygni/Data/'
fits_files = glob(datadir + 'hpolccd*r_hw.fits')
txt = glob(datadir + 'hpolccd*r_hw.fits.txt')
rfile = datadir+"/synthetic_filters/b_pcyg.txt"
r = pd.read_csv(rfile)
```

```python
class FObs(object):
    def __init__(self,w,q,u):
        self.w = w
        self.q = q
        self.u = u
        self.pol,self.pa = pc.polarization(self.q,self.u)
        self.amplitude = self.pol.max()-self.pol.min()
    
    def check_amp(self):
        fig,ax = plt.subplots(figsize=(15,10))
        ax.plot(self.w,self.pol)
        ax.axhline(self.pol.max())
        ax.axhline(self.pol.min())
        #ax.text(6560,1,str(self.amplitude))
        
        
```

```python
ws,qs,us = [],[],[]
for f in txt:
    w,f,q,u,e = pc.txt_QU_data(f, 1000, radial_velocity=-8.9)
    ws.append(w)
    qs.append(q)
    us.append(u)
```

```python
hw = [w[np.where((w >= 6540) & (w <= 6600))]    for w in ws]
hq = [q[np.where((w >= 6540) & (w <= 6600))]    for q,w in zip(qs,ws)]
hu = [u[np.where((w >= 6540) & (w <= 6600))]    for u,w in zip(us,ws)]
ow = [w[np.where((w >= 7500) & (w <= 7700))]    for w in ws]
oq = [q[np.where((w >= 7500) & (w <= 7700))]    for q,w in zip(qs,ws)]
ou = [u[np.where((w >= 7500) & (w <= 7700))]    for u,w in zip(us,ws)]
t = [pc.get_time(f)    for f in fits_files]
```

```python
amps = [FObs(w,q,u).amplitude    for w,q,u in zip(hw,hq,hu)]
oamps = [FObs(w,q,u).amplitude    for w,q,u in zip(ow,oq,ou)]
c=17
FObs(ow[c],oq[c],ou[c]).check_amp()
```

```python
fig,ax = plt.subplots(figsize=(15,10))
ax.plot(t,oamps/np.mean(oamps),label=r"amplitude of Oxygen feature")
ax.plot(t,amps/np.mean(amps),label=r"amplitude of Hydrogen feature")
#ax.plot(r.date,r.pol/r.pol.mean(),label="r-band polarization")
ax.set_xlim(49750,52600)
ax.legend()
```
